The left bar represents the profit generated from that sub-category.

# **Covariance matrix for Sales, Quantity, Discount, and Profit**

```python
# Check data types of columns in df
print(df1.dtypes)

# Select numeric columns only for covariance computation
numeric_cols = df1.select_dtypes(include=['float64', 'int64'])  # Adjust types as per your DataFrame

# Compute covariance matrix using selected numeric columns
cov_matrix = numeric_cols.cov()

# Create heatmap of the covariance matrix
fig, axes = plt.subplots(1, 1, figsize=(9, 6))
sns.heatmap(cov_matrix, annot=True)
plt.title('Heatmap of Covariance Matrix')
plt.show()
```

The heatmap shows the covariance matrix for Sales, Quantity, Discount, and Profit in the Sample Superstore dataset:

**Sales and Profit** have high positive covariance (7e+04), indicating that higher sales are associated with higher profits.

**Sales and Quantity** have a weak positive covariance (2.8e+02), suggesting a slight relationship between the number of items sold and sales.

**Discount and Profit** have a small negative covariance (-11), indicating that higher discounts might slightly reduce profits.

**Quantity and Profit** show a low positive covariance (35), meaning larger quantities sold slightly increase profits.

Overall, sales strongly correlate with profit, while discounts have a minimal negative effect on profit.

# **Relationship between sales and quantity for various product categories**

```python
# Example assuming 'Sales' and 'Quantity' are at indices [8, 9]
x = df1.iloc[:, [8, 9]].values

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

wcss = []

# Finding the optimal number of clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Plotting the clusters and centroids
sns.set_style("whitegrid")
sns.FacetGrid(df1, hue="Sub-Category", height=6).map(plt.scatter, 'Sales', 'Quantity')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.legend()
plt.show()
```

The scatter plot depicts the relationship between sales and quantity for various product categories in the "Sample Superstore" dataset. Each dot, color-coded by category, shows the sales amount (x-axis) versus the quantity sold (y-axis). Lower quantities (1-3) exhibit a broad range of sales, while higher quantities (10-14) generally correspond to lower sales. Notably, categories like Supplies and Machines achieve high sales even at lower quantities. Yellow dots represent centroids, indicating average points for clusters of data. This visualization highlights the diverse sales performance across different product categories.
